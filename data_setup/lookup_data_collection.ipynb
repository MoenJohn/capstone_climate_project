```python
import pandas as pd
```

```python
countries_df = pd.read_csv("../data/lookup_data/countries.csv")
countries_df.head()
```

```python
states_df = pd.read_csv("../data/lookup_data/states.csv")
states_df.head()
```

```python
colspecs = [(0, 11), (13, 20), (22, 30), (32, 37), (38, 40), (41, -1)]
stations_df = pd.read_fwf('../data/lookup_data/station_state_city.txt', colspecs=colspecs, names=['station_id', 'latitude', 'longitude', 'elevation', 'state_code', 'station_name'])
stations_df.head(100)
```

```python
stations_df.info()
```

```python
stations_lookup_df = pd.merge(stations_df, states_df, on=["state_code", "state_code"])
stations_lookup_df.head()
```
