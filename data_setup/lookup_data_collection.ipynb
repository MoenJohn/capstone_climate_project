```python
import pandas as pd
```

```python
countries_df = pd.read_csv("../data/lookup_data/countries.csv")
countries_df.head()
```

```python
states_df = pd.read_csv("../data/lookup_data/states.csv")
states_df.head()
```

```python
colspecs = [(0, 11), (13, 20), (22, 30), (32, 37), (38, 40), (41, -1)]
stations_df = pd.read_fwf('../data/lookup_data/station_state_city.txt', colspecs=colspecs, names=['station_id', 'latitude', 'longitude', 'elevation', 'state_code', 'station_name'])


stations_df.head(10)
```

```python
stations_df.info()
```

```python
stations_lookup_df = pd.merge(stations_df, states_df, on=["state_code", "state_code"])
stations_lookup_df.head()
```

```python
stations_lookup_df.info()
```

```python
df = pd.read_csv("../data/raw_station_data/2021.csv", names=["station_id", "date", "metric", "value", "a", "b", "c", "d"] )
df.head()
```

```python
df.info()
```

```python
df = df[df.metric == "TMAX"]
```

```python
df.info()
```

```python
df = df[df.station_id.str[:2] == "US"]
```

```python
df.info()
```

```python
df["year"] = df["date"].astype("str").str[:4]
df.head()
```

```python
df["month"] = df["date"].astype("str").str[4:6]
df.head()
```

```python
df.drop("Month", inplace=True, axis=1)
```

```python
df.head()
```

```python
df["day"] = df["date"].astype("str").str[6:8]
df.head()
```

```python
df.drop(["date","a","b","c","d"], axis=1, inplace=True)
```

```python
df["value"] = df.value / 10
```

```python
df.head()
```

```python
# df = df.join(stations_df, type="inner")
df = pd.merge(df, stations_df, how='inner', on = 'station_id')
```

```python
df.head()
```

```python
df = df.groupby(["state_code", "year", "month", "day"]).agg("max").reset_index()
```
